```python
#Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#Get the data
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
#What percentage of deals have undisclosed lenders?
len(df.lender[df.lender == '* Undisclosed']) / len(df.lender)
```

```python
grouped = df.groupby('lender')
```

```python
#Get series for number of deals and volume.
loan_amounts = grouped.agg(sum)['loan_amount']
num_deals = grouped.count()['property_id']

#Get numbers for LendingHome.
lh_loan_amount = loan_amounts['LendingHome']
lh_deals = num_deals['LendingHome']
```

```python
#Compare some basic statistics between lending home and all deals.
avg_deal_value = sum(loan_amounts) / sum(num_deals)
lh_avg_deal_value = lh_loan_amount / lh_deals

print("Average deal :" + str(avg_deal_value))
print("Average deal (LendingHome):" + str(lh_avg_deal_value))
```

```python
#remove the undisclosed values
loan_amounts_known_lenders = loan_amounts.drop("* Undisclosed").drop("* OTHER INSTITUTIONAL LEN")
num_deals_known_lenders = num_deals.drop("* Undisclosed").drop("* OTHER INSTITUTIONAL LEN")
```

```python
#Computes the same statistics as above.
avg_deal_value_known_lenders = sum(loan_amounts_known_lenders) / sum(num_deals_known_lenders)

print("Average deal (Only known lenders):" + str(avg_deal_value_known_lenders))
print("Average deal (LendingHome):" + str(lh_avg_deal_value))
```

```python
value_per_deal = loan_amounts_known_lenders / num_deals_known_lenders
value_per_deal_filtered = value_per_deal[num_deals_known_lenders > 10]
value_per_deal_filtered.max()
```

```python
#Draw the histogram of value per deal, with a marker for the average LendingHome deal.
plt.hist(value_per_deal_filtered, bins=50);
plt.scatter(lh_avg_deal_value, 0,s=50,c='red', marker='o');
plt.title("Average Deal Value");
```

```python
plt.hist(value_per_deal_filtered[value_per_deal_filtered < 2000001]);
plt.title("Loan Sizes (outliers excluded)");
```

```python
#Add a transaction_year column to the df to easily sort by dates.
df['transaction_year'] = list(map(lambda x: x.split('-')[0], df.transaction_date))
```

```python
grouped_lender_year = df.groupby(['lender', 'transaction_year'])
```

```python
lh_yearly_volume = grouped_lender_year.agg(sum)['loan_amount']['LendingHome']
lh_yearly_deals = grouped_lender_year.count()['property_id']['LendingHome']
```

```python
plt.plot(lh_yearly_volume[:len(lh_yearly_volume) - 1]);
plt.title("Loan Volume per year (Lending Home)");
```

```python
plt.plot(lh_yearly_deals[:len(lh_yearly_deals) - 1]);
plt.title("Number of Loans per year (Lending Home)");
```

```python
#Create new groupby object to easily aggregate by year.
df_known_lenders = df[df['loan_amount'] != 0.0]
grouped_year_lender = df_known_lenders.groupby(['transaction_year', 'lender'])
df_known_lenders
```

```python
def agg_multilevel_series(multi_series):
    """Returns the sum of values in a multilevel series, grouped by the first level index"""
    res = []
    curr_value = 0
    curr_index = multi_series.index[0][0]
    for pair in multi_series.index:
        value = multi_series.loc[pair]
        index = pair[0]
        if index != curr_index:
            res.append(curr_value)
            curr_value, curr_index = value, index
        else:
            curr_value += value
    res.append(curr_value)
    return res

yearly_sales = agg_multilevel_series(grouped_year_lender.agg(sum)['loan_amount'])
yearly_sales
```

```python
plt.plot(range(2012, 2017), yearly_sales[:len(yearly_sales) -1]);
plt.xticks(range(2012, 2017));
plt.title("Loan Volume per year");
```

```python
yearly_deals = agg_multilevel_series(grouped_year_lender.count()['property_id'])
yearly_deals
```

```python
plt.plot(range(2012, 2017), yearly_deals[:len(yearly_sales)-1]);
plt.title("Number of loans per year")
plt.xticks(range(2012, 2017));
```

```python
pd.Series(yearly_sales, index=range(2012, 2018)).pct_change()
```

```python
#computs average deal size on year to year basis, then converts to series
avg_size_yearly = [sales / deals for (sales, deals) in zip(yearly_sales, yearly_deals)]
avg_size_yearly = pd.Series(avg_size_yearly, index=range(2012, 2018))

#does the same calculation, but only for LendingHome
lh_avg_size_yearly = lh_yearly_volume / lh_yearly_deals
```

```python
plt.plot(avg_size_yearly.index, avg_size_yearly);
plt.title("Average Loan Size");
```

```python
plt.plot(lh_avg_size_yearly.index, lh_avg_size_yearly);
plt.title("Average Loan Size (Lending Home)");
```
